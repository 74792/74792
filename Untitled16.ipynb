Loading the Data

```python
import pandas as pd
import numpy as np

# Load the data from a CSV file
df = pd.read_csv('/content/Car Price Prediction.csv')

# Display the first few rows of the dataframe
print(df.head())
```

Inspecting the Data

```python
# Display general information about the dataframe
print(df.info())

# Statistical summary of the data
print(df.describe())

# Checking for missing values
print(df.isnull().sum())
```

Cleaning the Data

```python
# Dropping rows with missing values
df.dropna(inplace=True)

# Converting numeric columns if necessary
df['price'] = pd.to_numeric(df['price'], errors='coerce')


# Checking for outliers (example for 'price')
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'])
plt.title('Boxplot of Car Prices')
plt.show()
```

Descriptive Analysis of Variables

```python
# Distribution of fuel type
print(df['fueltype'].value_counts())

# Average price by fuel type
print(df.groupby('fueltype')['price'].mean())
```

Visualizations

```python
# Histogram of car prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Bar chart of the number of cars by fuel type
plt.figure(figsize=(10, 6))
sns.countplot(x='fueltype', data=df)
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()

# Relationship between engine size and price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='enginesize', y='price', data=df)
plt.title('Engine Size vs Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.show()
```

Building Model and Train the Model

```python
# Feature and target separation
X = df.drop('price', axis=1) # Features
Y = df['price'] # Target

from sklearn.model_selection import train_test_split


# Splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline

# Initializing OneHotEncoder
ohe = OneHotEncoder()

# List of categorical variables to be encoded
new_var = ['car_ID','symboling','CarName','fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','wheelbase','carlength','carwidth','carheight','curbweight','enginetype','cylindernumber','enginesize','fuelsystem','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']

# Fit OneHotEncoder to the categorical variables in the dataset
ohe.fit(X[new_var])

# Create a column transformer to apply OneHotEncoder to the categorical variables
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_), new_var), remainder = 'passthrough')

# Fit the pipeline on the training data
PL = make_pipeline(column_trans, LinearRegression())
PL.fit(X_train,  Y_train)
Y_pred = PL.predict(X_test)
r2_score(Y_test, Y_pred)

# Perform train-test split and model evaluation 1000 times with different random states
scores = []
for i in range(1000):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(X_train, Y_train)
    Y_pred = pipe.predict(X_test)
    scores.append(r2_score(Y_test, Y_pred))

scores[np.argmax(scores)]

# Perform train-test split using the best random state
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = np.argmax(scores))
lr = LinearRegression()
PL = make_pipeline(column_trans, lr)
PL.fit(X_train, Y_train)
Y_pred = PL.predict(X_test)
r2_score(Y_test, Y_pred)



```
